```python
import pandas
```

```python
with open("protein-aligned.txt") as file:
    info = file.read()
info = info.split(">")
while True:
    try:
        info.remove("")
    except:
        break
a_list, a_dictionary = [], {"Alphanumeric": [], "Protein": []}
for i in info:
    temp, seq = i.split("\n", maxsplit=1)
    a_dictionary["Alphanumeric"].append(temp.split(" ", maxsplit=1)[0])
    a_dictionary["Protein"].append("".join(seq.split("\n")))
```

```python
with open("BLOSUM62.txt") as file:
    info = file.read()
# split the info into separate lines and then each line into individual characters
info = info.split("\n")
info = [x.split() for x in info]
scores = pandas.DataFrame(info)
temp = list(scores.iloc[0])
temp[-1] = "-"
print(temp)
scores.columns, scores.index = temp, temp
# we drop the first row and first column as they are the headers
scores = scores.drop(scores.index[0], axis = 0)
scores = scores.drop(scores.columns[0], axis = 1)
scores
```

```python
distance_matrix = []
for i in range(len(a_dictionary["Protein"])):
    distance = []
    for j in range(len(a_dictionary["Protein"])):
        score = 0
        for x in range(len(a_dictionary["Protein"][j])):
            score += int(scores.at[a_dictionary["Protein"][i][x],a_dictionary["Protein"][j][x]])
        # we append the inverse (600 - score) (600 because it is more than max. value in the matrix)
        # because higher the score, more the similarity (easily observable from BLOSUM62) and thus,
        # we need to decrease it's value in the Pdistance matrix
        ans = (600-score)
        distance.append(ans)
    distance_matrix.append(distance)
```

```python
df = pandas.DataFrame(distance_matrix, index = a_dictionary['Alphanumeric'], columns = a_dictionary['Alphanumeric'])
df
```

```python
df.to_csv("Pdistance.txt")
```
