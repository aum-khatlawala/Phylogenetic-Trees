```python
def UPGMA(table, labels):
    while len(labels) >= 2:
        # first, we find the minimum value in the table
        minimum = float(10000)
        x, y = -1, -1
        for i in range(len(table)):
            for j in range(len(table[i])):
                if table[i][j] < minimum:
                    x, y = i, j
                    minimum = table[i][j]
        print(minimum)
        # then, we join the rows of the two indices which
        # contain the minimum value as per the UPGMA algorithm
        print(labels[x] + " and " + labels[y] + ": common ancestor at " + str(minimum/2) + " from the leaf node base of the tree")
        if y < x:
            x, y = y, x
        # we keep variables xcount and ycount to see what the weight of x and y is
        xcount=0
        ycount=0
        for i in labels[x]:
            # because we need to disregard the brackets
            if(i.isalpha()):
                xcount=xcount+1
        for i in labels[y]:
            # because we need to disregard the brackets
            if(i.isalpha()):
                ycount=ycount+1
        row = []
        # following the procedure shown in the tutorial to calculate
        # weighted average in every iteration
        for i in range(0, x):
            row.append((xcount*table[x][i] + ycount*table[y][i])/(xcount+ycount))
        table[x] = row
        for i in range(x+1, y):
            table[i][x] = (xcount*table[i][x]+ycount*table[y][i])/(xcount+ycount)
        for i in range(y+1, len(table)):
            table[i][x] = (xcount*table[i][x]+ycount*table[i][y])/(xcount+ycount)
            del table[i][y]
        del table[y]
        # finally, we join those two indices into the smaller index
        # and delete the larger valued index
        # absorption of the two entries with greatest similarity
        if y < x:
            x, y = y, x
        labels[x] = "(" + labels[x] + "," + labels[y] + ")"
        del labels[y]
        #print(labels)
        #print(xcount)
        #print(ycount)
        #print(table)
        #print("\n")
    return labels[0]
```
