```python
import pandas
```

```python
with open("nucleotide-aligned.txt") as file:
    info = file.read()
info = info.split(">")
while True:
    try:
        info.remove("")
    except:
        break
a_list, a_dictionary = [], {"Alphanumeric": [], "DNA": []}
for i in info:
    temp, seq = i.split("\n", maxsplit=1)
    a_dictionary["Alphanumeric"].append(temp.split(" ", maxsplit=1)[0])
    a_dictionary["DNA"].append("".join(seq.split("\n")))
```

```python
distance_matrix = []
for i in range(len(a_dictionary['DNA'])):
    distances = []
    for j in range(len(a_dictionary['DNA'])):
        mismatches = 0
        dashes = 0
        for x in range(len(a_dictionary['DNA'][i])):
            # because we need to remove the common dashes from the length to
            # get the effective length for the given pair.
            if a_dictionary['DNA'][i][x] == a_dictionary['DNA'][j][x] and a_dictionary['DNA'][i][x] == "-":
                dashes += 1
            # computing number of mismatches for the formula
            if a_dictionary['DNA'][i][x] != a_dictionary['DNA'][j][x]:
                mismatches += 1
        # computing effective length
        eff_length = len(a_dictionary['DNA'][i])-dashes
        # normalising the answer
        ans = mismatches/eff_length
        distances.append(ans)
    distance_matrix.append(distances)
```

```python
ndist = pandas.DataFrame(data=distance_matrix, index = a_dictionary['Alphanumeric'], columns = a_dictionary['Alphanumeric'])
ndist
```

```python
ndist.to_csv("./Ndistance.txt")
```
